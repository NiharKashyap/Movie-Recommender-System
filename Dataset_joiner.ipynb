```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("Datasets//movies_metadata.csv")
```

```python
ratings = pd.read_csv("Datasets//ratings.csv")
```

```python
movies
```

```python
movies = movies[['id', 'title']]
```

```python
movies.info()
```

```python
ratings.info()
```

```python
ratings = ratings.drop('timestamp', axis=1)
```

```python
ratings.rename(columns = {'movieId':'id'}, inplace=True)
```

```python
movies.dropna(inplace=True)
```

```python
movies['id'] = movies['id'].astype(np.int64)
```

```python
merged = pd.merge(ratings, movies, on='id')
```
